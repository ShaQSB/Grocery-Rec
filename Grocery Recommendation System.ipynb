```python
df1=df[['ProductID', 'ProductName', 'Brand', 'Price','Image_Url','Category']]
```

```python
df1
```

```python
df1['Tags']=df1['ProductName']+df['Category']
```

```python
df1
```

```python
cv = CountVectorizer(max_features=8208,stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(df1['Tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
similarity = cosine_similarity(vector)
```
